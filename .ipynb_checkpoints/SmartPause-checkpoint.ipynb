```python
face_detector_ = dlib.get_frontal_face_detector()
shape_predictor_ = dlib.shape_predictor("./models/shape_predictor_68_face_landmarks.dat")
def iterfaces(rgb):
        """Iterate over all detected faces"""
        for face in face_detector_(rgb, 1):
            yield face
            
def get_landmarks(rgb, face):
    return shape_predictor_(rgb, face)
```

```python
def function(img):
    for face in iterfaces(img):
        return face
```
